Describe what you see and interpret the the results: When viewing the head of the data frame, using the .head() function will print the first 5 observations of the dataframe along with the header row that tells you the name of the columns. The final thing shown in the head is the row number.

```python
# print summary statistics for the data set. Including numeric and categorical columns
# .describe() will show all numerical statistics, adding 'include='all'' ensures that both 
# numeric and categorical columns are used
iris.describe(include='all')
```

Describe what you see and interpret the the results: For the numeric columns, describe provides the average, standard deviation, minimum, maximum, and 25%, 50%, 75% quartile results for each column category. 

Since you obviously cannot take the average of a categorical column, the result for these statistics is 'NaN' or 'Not a Number.' For the categorical column, Species, the statistics provided include the number of unique results, the top result (in this case, we know they're actually evenly split, if we didn't we could tell because the the top result returns a frequency of 50, and there are 3 unique phrases among 150 total), and the frequency of the top result. 

Both numeric and categorical statistics return count. 

Nothing here particularly stands out or is useful in predicting the species of a flower since these are summary statistics for the entire data set.

```python
#Generate a box plot for the data set. This is only for the numeric data. 
iris.plot.box()
```

Describe what you see and interpret the the results: .plot.box() creates a box plot for all of the numeric data in the data set. The above graph matches what was shown in the summary statistics in the previous step. 

For example, looking at SepalLength, the average was 5.84, the red line in the graph above indicates the average and it falls just below 6. The maximum value was 7.9 and the minimum was 4.3, exactly where the top and bottom of the whisker lies. The box above encapsulates the 25% - 75% values, 5.1-6.4, where approximately 50% of the values fall. 

Since the data is now visualized, we can see that PetalLength has the greatest variation in the data and SepalWidth has the smallest variation. This may point to PetalLength being the best category to use in predicting a species, however, it's not conclusive as we are still looking at the data as a whole. 

```python
# Generate a line plot the dataset. This is only for numeric data
iris.plot.line()
```

Describe what you see and interpret the the results: As with the box plot, we can see that SepalWidth and Petal Length have the smallest and greatest variations within the data.

The one thing that begins to emerge, as suspected with the box plot, is that PetalLength may be the best category to use to predict species. From looking at the data earlier, we know that the data is listed be species in order. In other words, the different species are not mixed from one line to the next. After about 50, both PetalLength and PetalWidth experience a jump. After another 50, another disctinct jump is seen. As we start to break down the data by species, PetalLength and PetalWidth should be our starting points for predicting species.

```python
#Generate a histogram of the dataset
iris.plot.hist(bins = 15)
```

Describe what you see and interpret the the results: As with the previous two visualizations, the box plot shows similar results as SepalWidth spans the fewest bins and PetalLength spans the greatest. 

This view isn't quite as useful as the other two at looking at the data as whole as there isn't as much separation of the data compared to the line or boxplot. 

## Q 3

```python
#Print the summary statistics for each Species in the dataset
iris_species = pd.read_csv('iris.txt').groupby('Species')
iris_species.describe()
```

Describe what you see and interpret the the results: As inferred from the previous describe function called on all the data, you can see here that each Species is represented equally within the data set. 

There's nothing that stands out as odd in looking at these summary statistics, but you can see that the previous observation that there was a notable difference between the Species in Petal Length appears to be correct. 

```python
#Generate a boxplot for each species in the dataset. 
for group, name in iris_species:
    iris_species.get_group(group).plot.box(title=group)
```

Describe what you see and interpret the the results: As noticed previously, PetalLength and PetalWidth are offering the greatest differences between Species with the least amount of overlap. With Iris-Setosa, the Petal Length & Width stand out because there is not a lot of spread between the max & min values. This is in stark contrast to the other two species.

```python
#Generate a scatter matrix for the dataset 

scatter_matrix(iris, figsize=(6, 6))
```

Describe what you see and interpret the the results: This matrix allows us to see how each data point interacts with each other. For example, we can see that when Petal Length & Width are combined, there's a distinct grouping of points. This grouping could belong to one species. From this matrix, you can also see interaction between Petal Width and Sepal Width creating a distinct group as well.
